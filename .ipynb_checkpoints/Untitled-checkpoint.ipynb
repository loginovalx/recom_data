**ПУНКТ 2 - Train/test split**

```python
def train_test_split(
    sessions: list[list[int]],
) -> tuple[list[list[int]], list[int]]:
    """
    Разбиение сессий на train и test.

    Для каждой сессии:
      - все товары кроме последнего становятся
        обучающей сессией
      - последний товар становится тестовой целью

    Возвращаемые списки выровнены по индексу:
    test_targets[i] — товар, который нужно
    предсказать по train_sessions[i] как истории.

    Parameters
    ----------
    sessions : list of lists of ints
        Каждый вложенный список — одна сессия ID
        товаров, упорядоченная по времени.
        Все сессии содержат не менее 3 товаров.

    Returns
    -------
    train_sessions : list of lists of ints
        Сессии для обучения (исходные без последнего товара).
    test_targets : list of ints
        Следующий товар для предсказания по каждой сессии.
    """
    train_sessions = [session[:-1] for session in sessions]
    test_targets = [session[-1] for session in sessions]
    return train_sessions, test_targets
```

```python
import math
```

```python
#класс для создания графа и рассчета вероятности + обработка редких случаев
class ProbGraph:
    def __init__(self):
        self.graph = {}

    def _wilson_score(self, count, total):
        if total == 0:   #как будто не нужно
            return 0
        
        # z — это коэффициент уверенности. 
        # 1.96 соответствует 95% доверительному интервалу.
        z = 1.96
        
        # phat — это твоя "голая" вероятность (точечная оценка)
        phat = count / total
        
        # Магическая формула Вильсона (нижняя граница)
        denominator = 1 + z**2/total
        centre_point = phat + z**2/(2*total)
        delta = z * math.sqrt((phat * (1 - phat) + z**2 / (4 * total)) / total)
        
        # Возвращаем нижнюю границу (вычитаем delta)
        return (centre_point - delta) / denominator




        
    def create_graph(self, sessions):

        item_counts = {}
        for session in train_sessions:
            for i in range(len(session)):

                #считаем сколько встречается товар вообще, чтобы потом сделать топ товаров в test по кол-ву
                item = session[i]
                if item in item_counts:
                    item_counts[item] += 1
                else:
                    item_counts[item] = 1


                #реализация списка смежности графа
                if i < len(session) - 1:
                    first = session[i]
                    second = session[i+1]
            
                if first not in self.graph:
                    self.graph[first] = {}
                    
                if second in self.graph[first]:
                    self.graph[first][second] += 1
                else:
                    self.graph[first][second] = 1

        #создали словарь, где будут 10 самых популярных item_counts
        self.global_top = sorted(item_counts, key=item_counts.get, reverse=True)[:10]


        
    
    def get_probability(self, id1, id2,total=None, use_wilson=True):
        if id1 not in self.graph:
            return 0.0
        
        count = self.graph[id1].get(id2, 0) #на 0 не проверяем, так как первый if это учитывает
        if total is None:
            total = sum(self.graph[id1].values())
            
        if use_wilson:
            return self._wilson_score(count, total)
        else:
            return count / total


  
    # def predict(self, current_item,use_wilson=False):
    #     """
    #     Формирует финальный топ-10 рекомендаций для конкретного товара
    #     """
    #     recommendations = []

    #     # 1. Если товар есть в графе, достаем его соседей
    #     if current_item in self.graph:
    #         #словарь, куда может переходить
    #         neighbors = self.graph[current_item]
    #         total = sum(neighbors.values())
            
    #         # Собираем список (товар, вероятность_с_Вильсоном)
    #         scored_neighbors = []
    #         for neighbor_id in neighbors:
    #             score = self.get_probability(current_item, neighbor_id, total, use_wilson=use_wilson)
    #             scored_neighbors.append((neighbor_id, score))
            
    #         # Сортируем по убыванию "умной" вероятности
    #         scored_neighbors.sort(key=lambda x: x[1], reverse=True)
            
    #         # Берем ID товаров (только те, что не являются текущим)

    #         recommendations = [item[0] for item in scored_neighbors][:10]

    #     # 3. ДОБАВКА ПОПУЛЯРНЫХ (Fallback):
    #     # Если рекомендаций меньше 10 (или вообще 0), берем товары из global_top
    #     if len(recommendations) < 10:
    #         for top_item in self.global_top:
    #             # Добавляем товар, если его еще нет в списке и это не сам текущий товар
    #             if top_item not in recommendations and top_item != current_item:
    #                 recommendations.append(top_item)
                
    #             # Как только набрали 10 штук — выходим
    #             if len(recommendations) == 10:
    #                 break
        
    #     return recommendations

    def predict(self, session, use_wilson=True):
        """
        Формирует топ-10 на основе двух последних товаров сессии
        """
        # Если сессия пустая - отдаем просто глобальный топ
        if not session:
            return self.global_top

        # Определяем веса для истории (экспериментально: последний важнее)
        weights = {
            'last': 1.0,
            'prev': 0.5
        }
        
        # Словарь для агрегации баллов: {item_id: total_weighted_score}
        candidate_scores = {}
        
        # 1. СМОТРИМ ПОСЛЕДНИЙ ТОВАР
        last_item = session[-1]
        if last_item in self.graph:
            neighbors = self.graph[last_item]
            total = sum(neighbors.values())
            for nid, count in neighbors.items():
                score = self.get_probability(last_item, nid, total, use_wilson=use_wilson)
                candidate_scores[nid] = candidate_scores.get(nid, 0) + (score * weights['last'])

        # 2. СМОТРИМ ПРЕДПОСЛЕДНИЙ ТОВАР (если он есть)
        if len(session) > 1:
            prev_item = session[-2]
            if prev_item in self.graph:
                neighbors = self.graph[prev_item]
                total = sum(neighbors.values())
                for nid, count in neighbors.items():
                    score = self.get_probability(prev_item, nid, total, use_wilson=use_wilson)
                    # Прибавляем к уже имеющемуся баллу или создаем новый
                    candidate_scores[nid] = candidate_scores.get(nid, 0) + (score * weights['prev'])

        # 3. РАНЖИРОВАНИЕ И ФИЛЬТРАЦИЯ
        # Сортируем кандидатов по накопленному баллу
        sorted_candidates = sorted(candidate_scores.items(), key=lambda x: x[1], reverse=True)
        
        # Фильтруем: исключаем товары, которые уже ЕСТЬ в текущей сессии
        # (Пользователь их уже видел, зачем предлагать снова?)
        session_set = set(session)
        recommendations = [item[0] for item in sorted_candidates if item[0] not in session_set][:5]

        # 4. ДОБАВКА ПОПУЛЯРНЫХ (Fallback)
        if len(recommendations) < 10:
            for top_item in self.global_top:
                if top_item not in recommendations and top_item not in session_set:
                    recommendations.append(top_item)
                if len(recommendations) == 10:
                    break
        
        return recommendations
        
```

```python
def hit_at_k(
    recommendations: list[list[int]],
    true_items: list[int],
    k: int = 10,
) -> float:
    """
    Вычисление Hit@K для списка предсказаний.

    Parameters
    ----------
    recommendations : list of lists of ints
        recommendations[i] — ранжированный список
        рекомендаций для i-го примера.
    true_items : list of ints
        true_items[i] — истинный следующий товар
        для i-го примера.
    k : int
        Отсечка top-K (по умолчанию 10).

    Returns
    -------
    float
        Hit@K, значение от 0 до 1.
    """
    assert len(recommendations) == len(true_items), \
        "recommendations и true_items должны совпадать по длине"

    hits = 0
    for recs, true_item in zip(recommendations, true_items):
        if true_item in recs[:k]:
            hits += 1

    return hits / len(true_items)
```

```python
# --- 1. Обучаем модель на тренировочных сессиях ---
model = ProbGraph()
model.create_graph(train_sessions)

# --- 2. Получаем предсказания для ТЕСТОВОЙ выборки ---
model_recommendations = []
for session in train_sessions:
    # Предсказываем по последнему товару каждой обучающей сессии
    last_item = session[-1] 
    res = model.predict(last_item)
    model_recommendations.append(res)

# --- 3. Считаем бейзлайн (10 самых популярных товаров всегда) ---
# Просто дублируем глобальный топ для каждой сессии
baseline_recommendations = [model.global_top for _ in range(len(train_sessions))]

# --- 4. Оцениваем качество с помощью функции hit_at_k ---
# (Убедись, что функция hit_at_k объявлена в твоем коде)
model_score = hit_at_k(model_recommendations, test_targets, k=10)
baseline_score = hit_at_k(baseline_recommendations, test_targets, k=10)

print(f"Hit@10 Модели: {model_score:.4f}")
print(f"Hit@10 Бейзлайна (Top Pop): {baseline_score:.4f}")

if model_score > baseline_score:
    print("Результат: Модель успешно обошла бейзлайн")
else:
    print("Результат: Бейзлайн оказался сильнее. Нужно проверить данные")
```

```python
# --- 1. Обучаем модель (оставляем как было) ---
model = ProbGraph()
model.create_graph(train_sessions)

# --- 2. Получаем предсказания для ТЕСТОВОЙ выборки ---
model_recommendations = []

# ВАЖНО: Мы идем по сессиям, которые соответствуют нашим test_targets
for session in train_sessions:
    # ИЗМЕНЕНИЕ: Передаем в predict ВСЮ сессию (список), а не только last_item
    # Также явно указываем use_wilson=True, чтобы почувствовать разницу
    res = model.predict(session, use_wilson=True)
    model_recommendations.append(res)

# --- 3. Считаем бейзлайн (оставляем как было) ---
baseline_recommendations = [model.global_top for _ in range(len(train_sessions))]

# --- 4. Оцениваем качество ---
model_score = hit_at_k(model_recommendations, test_targets, k=10)
baseline_score = hit_at_k(baseline_recommendations, test_targets, k=10)

print(f"Hit@10 Модели (2-step + Wilson): {model_score:.4f}")
print(f"Hit@10 Бейзлайна (Top Pop): {baseline_score:.4f}")
```
